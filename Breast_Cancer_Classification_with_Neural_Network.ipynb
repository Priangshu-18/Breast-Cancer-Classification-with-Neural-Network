```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_data)
```

```python
data_frame = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
```

```python
print(data_frame)
```

```python
data_frame['label'] = breast_cancer_data.target
```

```python
data_frame.tail()
```

```python
data_frame.shape
```

```python
data_frame.info()
```

```python
data_frame.describe()
```

```python
data_frame['label'].value_counts()
```

*   1 -> Benign
*   0 -> Malignant

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

**Building the Neural Network**

Importing the tensorflow and keras

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Setting up the layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiling the neural network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

**Sparse Categorical Crossentropy Label Encoding**

Dog = 0

Cat = 1

Horse = 2




**One-Hot Encoding Labels**

Dog = [1 0 0]

Cat = [0 1 0]

Horse = [0 0 1]

```python
# Training the neural network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing the accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')

plt.legend(['Training Data', 'Validation Data'], loc = 'lower right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
print(X_test_std.shape)
print(X_test_std[0])
```

```python
Y_pred = model.predict(X_test_std)
```

```python
print(Y_pred.shape)
print(Y_pred[0])
```

model.predict() gives the prediction probability of each class for that data

```python
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```
